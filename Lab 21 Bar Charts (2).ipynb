### Task 1: Visualizing Data Distributions


##### 1. Histogram of `ConvertedCompYearly`


Visualize the distribution of yearly compensation (`ConvertedCompYearly`) using a histogram.



```python
## Write your code here
df = df.dropna(subset='ConvertedCompYearly')
plt.figure(figsize=(12, 6)) 
plt.hist(df['ConvertedCompYearly'], bins=30, color="blue", edgecolor="black") 
plt.title('The Distribution Of Yearly Compensation') 
plt.xlabel('Yearly Compensation') 
plt.ylabel('Frequency')
plt.grid(True) 
plt.show()
```

##### 2. Box Plot of `Age`


Since `Age` is categorical in the dataset, convert it to numerical values for a box plot.



```python
## Write your code here
age_mapping = {'Under 18 years old':17, '35-44 years old':40, '45-54 years old':50,
 '18-24 years old':21, '25-34 years old':30, '55-64 years old':60, 'Prefer not to say':0,
 '65 years or older':70}
df["Age"]= df["Age"].map(age_mapping)
index_to_drop = df[df["Age"] == 0].index 
df = df.drop(index_to_drop) 
plt.figure(figsize=(12, 8)) 
plt.boxplot(df['Age'], vert=False)
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.yticks([])
plt.grid(True)
plt.show()
```

### Task 2: Visualizing Relationships in Data


##### 1. Scatter Plot of `Age_numeric` and `ConvertedCompYearly`


Explore the relationship between age and compensation.



```python
## Write your code here
plt.figure(figsize=(12, 8)) 
plt.scatter(df['Age'], df['ConvertedCompYearly'], marker='o', edgecolors='blue', linewidths=2, alpha=0.5) 
plt.title('Relationship Between Age and Compensation')
plt.xlabel('Age')
plt.ylabel('Yearly Compensation')
plt.grid(True)
plt.show()
```

##### 2. Bubble Plot of `ConvertedCompYearly` and `JobSatPoints_6` with `Age_numeric` as Bubble Size


Explore how compensation and job satisfaction are related, with age as the bubble size.


```python
## Write your code here
df = df.dropna(subset='JobSatPoints_6') 
plt.figure(figsize=(12, 8))
plt.scatter(df['ConvertedCompYearly'], df['JobSatPoints_6'], s=df['Age']*2, c=df['Age'], alpha=0.6, cmap='viridis') 
plt.colorbar(label='Age')
plt.title('Compensation vs Job Satisfaction (Bubble Size by Age)')
plt.xlabel('Yearly Compensation')
plt.ylabel('Job Satisfaction Points (6)')
plt.grid(True)
plt.show()
```

### Task 3: Visualizing Composition of Data with Bar Charts


##### 1. Horizontal Bar Chart of `MainBranch` Distribution


Visualize the distribution of respondents’ primary roles to understand their professional focus.



```python
## Write your code here
main_branch_counts = df['MainBranch'].value_counts()
plt.figure(figsize=(10, 6))
main_branch_counts.plot(kind='barh')
plt.title('Distribution of Respondents by Main Branch')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.grid(True)
plt.show()
```

##### 2. Vertical Bar Chart of Top 5 Programming Languages Respondents Want to Work With


Identify the most desired programming languages based on `LanguageWantToWorkWith`.



```python
## Write your code here
languages=df['LanguageHaveWorkedWith'].str.split(';').explode().str.strip()
top_languages= languages.value_counts().head(10)
plt.figure(figsize=(12, 8)) 
top_languages.plot(kind='bar')
plt.title('Top 10 Programming Languages Respondents have to Work With')
plt.xlabel('Programming Language')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

##### 3. Stacked Bar Chart of Median `JobSatPoints_6` and `JobSatPoints_7` by Age Group


Compare job satisfaction metrics across different age groups with a stacked bar chart.


```python
df= df.dropna(subset=['JobSatPoints_6','JobSatPoints_7']) 
bins=[20, 30, 40, 50, 60]
labels=['20-29', '30-39', '40-49', '50-59'] 
df['age_group'] = pd.cut(df['Age'], bins= bins, labels=labels).copy() 

job_sat_by_age = df.groupby('age_group', observed=True)[['JobSatPoints_6', 'JobSatPoints_7']].median()
plt.figure(figsize=(12, 7))
job_sat_by_age.plot(kind='bar', stacked=True)
plt.title('Median Job Satisfaction Points by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Job Satisfaction Points')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Job Satisfaction Metric')
plt.show()
```

##### 4. Bar Chart of Database Popularity (`DatabaseHaveWorkedWith`)


Identify the most commonly used databases among respondents by visualizing `DatabaseHaveWorkedWith`.



```python
## Write your code here
db = df['DatabaseWantToWorkWith'].str.split(';').explode().str.strip()
db_common = db.value_counts().head(10) 
plt.figure(figsize=(12, 7)) 
db_common.plot(kind='bar') 
plt.title('Top 10 Databases Respondents want to Work With')
plt.xlabel('Database')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### Task 4: Visualizing Comparison of Data with Bar Charts


##### 1. Grouped Bar Chart of Median `ConvertedCompYearly` for Different Age Groups


Compare median compensation across multiple age groups with a grouped bar chart.



```python
## Write your code here
median_comp_by_age = df.groupby('age_group', observed=True)['ConvertedCompYearly'].median()
plt.figure(figsize=(12, 7))
median_comp_by_age.plot(kind='bar')
plt.title('Median Yearly Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Yearly Compensation')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

##### 2. Bar Chart of Respondent Count by Country


Show the distribution of respondents by country to see which regions are most represented.



```python
## Write your code here
top_countries = df['Country'].value_counts().head(10)
plt.figure(figsize=(12, 7))
top_countries.plot(kind='bar')
plt.title('Top 10 Countries by Respondent Count')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
