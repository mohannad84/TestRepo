### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
print("name of columns:")
print(df.columns.tolist())
print("\n Data information :")
print(df.info())
print("\n Descriptive statistics:")
print(df.describe(include="all"))
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
missing_values = df.isnull().sum() 
print("Missing values in each column") 
print(missing_values[missing_values > 0])
cols_to_check = ["Employment", "JobSat", "YearsCodePro"] 
df["YearsCodePro"] = pd.to_numeric(df["YearsCodePro"], errors='coerce')
df_cleaned = df.dropna(subset=cols_to_check) 
print("\n Number of rows before and after cleaning :") 
print("before", len(df)) 
print("after" , len(df_cleaned))
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
print("\n Employment status distribution:") 
print(df_cleaned["Employment"].value_counts())
print("\n Distribution of job satisfaction:") 
print(df_cleaned["JobSat"].value_counts()) 
print("\n Years of experience statistics:")
print(df_cleaned["YearsCodePro"].describe())
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
plt.figure(figsize=(10, 6)) 
sns.kdeplot(data=df_cleaned, x="JobSat", fill=True) 
plt.title("Job satisfaction density distribution") 
plt.xlabel("Satisfaction level (0-10)") 
plt.ylabel("Density") 
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
languages_worked = df_cleaned["LanguageHaveWorkedWith"].str.split(";").explode().value_counts().head(10) 
languages_wanted = df_cleaned["LanguageWantToWorkWith"].str.split(";").explode().value_counts().head(10) 
plt.figure(figsize=(12, 6)) 
sns.barplot(x=languages_worked.values, y= languages_worked.index, color="blue", label= "Currently working there") 
sns.barplot(x=languages_wanted.values, y=languages_wanted.index, color="orange", label="He wants to work there")  
plt.title("Comparison of the top 10 programming languages")
plt.xlabel("number")
plt.legend()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
# Create a countplot for RemoteWork by Region
plt.figure(figsize=(15, 9))
sns.countplot(data=df_cleaned, x='RemoteWork', hue='Country', palette='viridis')
plt.title('Remote Work Preferences by Region')
plt.xlabel('Remote Work Preference')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
!pip install scipy
from scipy.stats import spearmanr 
corr, p_value = spearmanr(df_cleaned["JobSat"], df_cleaned["YearsCodePro"]) 
print(f"Spearman  correlation coefficient between job satisfaction and years of experience : {corr:.3f}") 
print(f" p value : {p_value:.3f}") 
fig, ax = plt.subplots(figsize=(10, 6))
sns.regplot(x="YearsCodePro", y="JobSat", data=df_cleaned, scatter_kws={"alpha":0.3}, line_kws={"color":"red", "linewidth":2}, ax=ax)
ax.set_title("The relationship between job satisfaction and years of programming experience") 
ax.set_xlabel("Years of programming experience") 
ax.set_ylabel("job satisfaction") 
plt.show()
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
cross_tab = pd.crosstab(df_cleaned["Employment"], df_cleaned["EdLevel"], normalize="index") * 100 
print("Employment status vs. educational level (% within each employment status):") 
print(cross_tab) 
cross_tab.plot(kind="bar", stacked=True, figsize=(12, 6)) 
plt.title('Distribution of educational level by job status')
plt.xlabel('Employment status')
plt.ylabel('percentage')
plt.legend(title='Educational level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
